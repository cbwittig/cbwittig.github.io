## Generating ngrams

```python
def ngrams(input, n):
    input = input.split(' ')
    output = []
    for i in range(len(input)-n+1):
        output.append(input[i:i+n])
    return output
```

```python
# bigram on pre-tokenized text - error on post-tokenized -- NEED TO INVESTIGATE

bigram = ngrams(file_text, 2)
bigram
```
