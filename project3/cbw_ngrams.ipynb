```python
#Import SINGLE ngram file directly from txt file
df = pd.read_csv("journal-article-10.2307_20862402-ngram2.txt", header=None)
df.head()
```

```python
 #Create new data frame with split value columns - word count by count - splitting on /t
new = df[0].str.split("\t", n=1, expand = True) 
new.head()
  
```

```python
#Search for terms with column
mterm_freq = named[named['ngram2'].str.match('writ') | named['ngram2'].str.match('literac')]
mterm_freq
```

```python
#Reading multiple files in - retain file name for referencing

articles = glob.glob('ngram_test/*.txt')
d = list()
for article_file in articles:
    with open(article_file, encoding='utf-8') as f:
        article = os.path.basename(article_file.split('.')[0])
        d.append(pd.DataFrame({'article': article, 'lines': f.readlines()}))
doc = pd.concat(d)
doc.head()
```
